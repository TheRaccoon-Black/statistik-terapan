```python
#import library
import numpy as np
import pandas as pd
#import dataset
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vR49WPBvyvBlfNo1wEwE6epoXKrpHfcLhAkI87s3vB2OSctIgkp7pxQhg-IX3LHK_AQmgOlpmZpnebw/pub?gid=382860817&single=true&output=csv')
#menampilkan 5 data teratas
df.head(5)
```

```python
#menampilkan informasi dari dataset
df.info()
```

```python
#menampilkan informasi missing values
print("Banyaknya Missing Values : \n")
df.isnull().sum()
```

```python
#handling missing values dengan mean
df['Cars'].fillna(df['Cars'].mean(), inplace = True)
df['Age'].fillna(df['Age'].mean(), inplace = True)

#handling missing values dengan median
df['Income'].fillna(df['Income'].median(), inplace = True)
df['Children'].fillna(df['Children'].median(), inplace = True)

#handling missing value dengan modus
df['Marital Status'].fillna(df['Marital Status'].mode()[0], inplace = True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace = True)
df['Home Owner'].fillna(df['Home Owner'].mode()[0], inplace = True)


#menampilkan informasi missing values
print("Banyaknya Missing Values : \n")
df.isnull().sum()
```
